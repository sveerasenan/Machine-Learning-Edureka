```python
#1. Hurricanes graph
import pandas as pd
import numpy as  np
from matplotlib import pyplot as plt
hurricanesdf = pd.read_csv("Hurricanes.csv")
print (hurricanesdf)
plt.bar(hurricanesdf["Year"],hurricanesdf["Hurricanes"],color="green")
plt.title("Hurricanes Occurring in the US along the cost of the Atlantic")
plt.ylabel("Number of Hurricanes")
plt.xlabel("Year")
plt.show()
```

```python
#2.City Temperature graph
import pandas as pd
from matplotlib import pyplot as plt
citydf = pd.read_csv("CityTemps.csv")
#bins = [0,5,10,15,20,25,30]
print (citydf)
plt.hist(citydf["Moscow"],bins=10,histtype="bar",rwidth=0.5,color='red')
plt.hist(citydf["San Francisco"],bins=10,histtype="bar",rwidth=0.5,color='blue')
plt.legend(['Moscow','San Francisco'])
plt.show()
        
```

```python
#5.9 scatter plot
import pandas as pd
from matplotlib import pyplot as plt
empDetails = {'first_name':['Jason','Molly','Tina','Jake','Amy'],'last_name':['Miller','Jacobson','Ali','Milner','Cooze'],'female':[0,1,1,0,1],'age':[42,52,36,24,73],'preTestScore':[4,24,31,2,3],'postTestScore':[25,94,57,62,70]}
scatterdf = pd.DataFrame(empDetails)

plt.scatter(scatterdf['age'],scatterdf['preTestScore'],marker='^')
plt.scatter(scatterdf['age'],scatterdf['postTestScore'],color='red')
plt.legend(['Pre Test Score','Post Test SCore'])
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Score")
plt.show()

#plot scores colored by sex
pretestscorefemale = scatterdf.loc[scatterdf["female"] == 1]
pretestscoremale = scatterdf.loc[scatterdf["female"] == 0]
posttestscorefemale = scatterdf.loc[scatterdf["female"] == 1]
posttestscoremale = scatterdf.loc[scatterdf["female"] == 0]

plt.scatter(pretestscorefemale['age'],pretestscorefemale['preTestScore'],color='pink',marker='^')
plt.scatter(posttestscorefemale['age'],posttestscorefemale['postTestScore'],color='pink')            
plt.scatter(pretestscoremale['age'],pretestscoremale['preTestScore'],color='blue',marker='^')
plt.scatter(posttestscoremale['age'],posttestscoremale['postTestScore'],color='blue')  
plt.legend(['Pre Test Score','Post Test SCore'])
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Score")
plt.show()
```
