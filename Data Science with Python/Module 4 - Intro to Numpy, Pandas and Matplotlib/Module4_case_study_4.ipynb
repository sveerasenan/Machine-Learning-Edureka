```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i])
        
bigmartdf = pd.read_csv("BigMartSalesData.csv")
sales_2011 = bigmartdf.loc[bigmartdf["Year"] == 2011]

mothly_sales = sales_2011.groupby("Month")["Amount"].sum()
#1. Plot total sales per month
plt.plot(mothly_sales.index,mothly_sales.values)
plt.title("2011 Plot")
plt.xlabel("Month")
plt.ylabel("Sale Amount")
plt.show()
#Feburary has the lowest sales

#2. Plot total sales per month as bar chart
plt.bar(mothly_sales.index,mothly_sales.values)
# calling the function to add value labels
addlabels(mothly_sales.index,mothly_sales.values)
plt.title("2011 Plot")
plt.xlabel("Month")
plt.ylabel("Sale Amount")
plt.show()

#3. Country wise pie chart. Which country contributes the highest sales
#UK contributes the highest sales
country_sales = sales_2011.groupby("Country")["Amount"].sum()
plt.pie(country_sales.values,labels=country_sales.index,shadow=True, startangle=90)
plt.show()

#4.scatter plot for the invoice amount
filtered_data = bigmartdf[bigmartdf["Amount"] <= 100]
sample_data = filtered_data.sample(n=1000, random_state=42)
plt.scatter(sample_data["InvoiceNo"], sample_data["Amount"], alpha=0.5, color = "red")
#More concentration is between 0-20
plt.show()
print (bigmartdf)
```
