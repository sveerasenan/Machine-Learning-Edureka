```python
import pandas as pd
from matplotlib import pyplot as plt
moviesdf = pd.read_csv("C:\\Users\\sveer\\Edureka\\Data Science\\Assignments\\Module 5\\HollywoodMovies.csv")
questdf = moviesdf[(moviesdf["Story"] == "Quest")]

#1.Highest rated movie
sortedquestdf = (questdf.sort_values(by='RottenTomatoes',ascending=False))
print ("Highest rated movie")
print (sortedquestdf[:1])

#2. Genre which has the greatest number of movie releases
groupedgenredf = moviesdf.groupby("Genre").count()
sortedgenredf = (groupedgenredf.sort_values(by='Movie',ascending=False))
print ("Greatest number of movie releases")
print (sortedgenredf[:1])

#3. Top five costliest budgets
sortedcostlydf = (moviesdf.sort_values(by='Budget',ascending=False))
print ("Top five costliest budgets")
print (sortedcostlydf[:5])

#4. correspondence between critics evaluation of a movie and acceptance by the public
plt.scatter(moviesdf["RottenTomatoes"],moviesdf["Profitability"])
plt.ylim(0,4000)
plt.show()
#There is no correspondence.
```
