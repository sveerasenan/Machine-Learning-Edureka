```python
import pandas as pd
salarydf = pd.read_csv("C:\\Users\\sveer\\Edureka\\Data Science\\Assignments\\Module 5\\Salaries.csv", dtype={12: str})
print (salarydf.info())

#1. How much salary cost has increased from the year 2011 to 2014
df2011 = salarydf.loc[salarydf["Year"] == 2011]
df2014 = salarydf.loc[salarydf["Year"] == 2014]
print ("Total salary in 2011", df2011["TotalPay"].sum())
print ("Total salary in 2014", df2014["TotalPay"].sum())
print ("Increased cost",df2014["TotalPay"].sum() - df2011["TotalPay"].sum())

#2. Which job title in the year 2014 has the highest mean salary?
meandf = df2014.groupby("JobTitle")["TotalPay"].mean()
meandf.sort_values(ascending=False,inplace=True)
print (meandf)
print ("highest mean job title",meandf.index[0])

#3. How much money could have been saved in the Year 2014 by stopping OverTimePay
print("Overtimepay for 2014",df2014["OvertimePay"].sum())

#4. Top 5 common jobs in 2014 and how much do they cost SFO
print ("Common jobs")
common_jobs = df2014['JobTitle'].value_counts().head(5)
top_jobs_df = df2014[df2014['JobTitle'].isin(common_jobs.index)]
total_salary_costs = top_jobs_df.groupby('JobTitle')['TotalPay'].sum()
total_salary_costs_formatted = total_salary_costs.apply(lambda x: "{:,.2f}".format(x))
print (total_salary_costs_formatted)

#5. Top earning employee across all years
salarydfsorted = salarydf.sort_values(by='TotalPay',ascending=False)
print ("Top earning employee",salarydfsorted.iloc[0])

#e2. In year 2011 OverTimePay was what percentage of  TotalPayBenefits
totalpay2011 = df2011["TotalPayBenefits"].sum()
overtimepay2011 = df2011["OvertimePay"].sum()
print ("TotalPayBenefits pay was over by ", (overtimepay2011/totalpay2011)*100)

#e3. Which Job Title in the Year 2014 has the lowest mean salary
salarymean = df2014.groupby("JobTitle")["TotalPayBenefits"].mean()
sortedsal = salarymean.sort_values(ascending=True)

# The job title with the lowest mean salary
lowest_salary_job = sortedsal.index[0]
lowest_salary_value = sortedsal.iloc[0]

print(f"The job title with the lowest mean salary in 2014 is '{lowest_salary_job}' with a mean salary of {lowest_salary_value:.2f}")
```
